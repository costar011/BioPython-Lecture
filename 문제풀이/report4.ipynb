## ğŸ’¡ ë¬¸ì œ ì„¤ëª…

ì •ìˆ˜ `l`ê³¼ `r`ì´ ì£¼ì–´ì¡Œì„ ë•Œ, `l` ì´ìƒ `r` ì´í•˜ì˜ ì •ìˆ˜ ì¤‘ì—ì„œ ìˆ«ì **"0"ê³¼ "5"ë¡œë§Œ ì´ë£¨ì–´ì§„** ëª¨ë“  ì •ìˆ˜ë¥¼ ì˜¤ë¦„ì°¨ìˆœìœ¼ë¡œ ì €ì¥í•œ ë°°ì—´ì„ return í•˜ëŠ” `solution` í•¨ìˆ˜ë¥¼ ì™„ì„±í•´ ì£¼ì„¸ìš”.

ë§Œì•½ ê·¸ëŸ¬í•œ ì •ìˆ˜ê°€ ì—†ë‹¤ë©´, `[-1]`ì´ ë‹´ê¸´ ë°°ì—´ì„ return í•©ë‹ˆë‹¤.

| :---: | :---: | :--- |

### ì…ì¶œë ¥ ì˜ˆ ì„¤ëª…

* **ì…ì¶œë ¥ ì˜ˆ #1:**
  5 ì´ìƒ 555 ì´í•˜ì˜ 0ê³¼ 5ë¡œë§Œ ì´ë£¨ì–´ì§„ ì •ìˆ˜ëŠ” ì‘ì€ ìˆ˜ë¶€í„° 5, 50, 55, 500, 505, 550, 555ê°€ ìˆìŠµë‹ˆë‹¤. ë”°ë¼ì„œ `[5, 50, 55, 500, 505, 550, 555]`ë¥¼ return í•©ë‹ˆë‹¤.

* **ì…ì¶œë ¥ ì˜ˆ #2:**
  10 ì´ìƒ 20 ì´í•˜ì´ë©´ì„œ 0ê³¼ 5ë¡œë§Œ ì´ë£¨ì–´ì§„ ì •ìˆ˜ëŠ” ì—†ìŠµë‹ˆë‹¤. ë”°ë¼ì„œ `[-1]`ì„ return í•©ë‹ˆë‹¤.

```python
def solution1(l, r):
    answer = []
    valid_chars = {'0', '5'}

    for num in range(l, r + 1):
        if set(str(num)).issubset(valid_chars):
            answer.append(num)
    return answer if answer else [-1]

# ----------------------------------------------------------------------
## ğŸ“ ì…ì¶œë ¥ ì˜ˆì‹œ ì±„ì  êµ¬ì¡°

# ğŸ“ ì…ì¶œë ¥ ì˜ˆì‹œ ë°ì´í„°
# (ì…ë ¥ê°’: l, r, ê¸°ëŒ€ê°’: result)
test_cases = [
    # í”„ë¡œê·¸ë˜ë¨¸ìŠ¤ ê¸°ë³¸ ì…ì¶œë ¥ ì˜ˆì‹œ #1
    (5, 555, [5, 50, 55, 500, 505, 550, 555]),

    # í”„ë¡œê·¸ë˜ë¨¸ìŠ¤ ê¸°ë³¸ ì…ì¶œë ¥ ì˜ˆì‹œ #2
    (10, 20, [-1]),

    # ì¶”ê°€ í…ŒìŠ¤íŠ¸ ì¼€ì´ìŠ¤ 1: 5ë§Œ í¬í•¨
    (1, 5, [5]),

    # ì¶”ê°€ í…ŒìŠ¤íŠ¸ ì¼€ì´ìŠ¤ 2: 500000 ì´ìƒì˜ í° ë²”ìœ„
    (500000, 500005, [500000, 500005]),

    # ì¶”ê°€ í…ŒìŠ¤íŠ¸ ì¼€ì´ìŠ¤ 3: ì•„ë¬´ê²ƒë„ ì—†ì„ ë•Œ
    (100, 499, [-1]),
]

def run_tests(test_cases):
    """ì£¼ì–´ì§„ ì…ì¶œë ¥ ì˜ˆì‹œë¡œ solution1 í•¨ìˆ˜ë¥¼ í…ŒìŠ¤íŠ¸í•˜ê³  ê²°ê³¼ë¥¼ ì¶œë ¥í•©ë‹ˆë‹¤."""
    print("--- ğŸ”¢ í”„ë¡œê·¸ë˜ë¨¸ìŠ¤ 'ë°°ì—´ ë§Œë“¤ê¸° 2' í…ŒìŠ¤íŠ¸ ì‹¤í–‰ ---")

    all_passed = True

    for l, r, expected_result in test_cases:
        actual_result = solution1(l, r)

        # ê²°ê³¼ ë¹„êµ ë° ì¶œë ¥
        if actual_result == expected_result:
            status = "âœ… í†µê³¼ (PASS)"
        else:
            status = "âŒ ì‹¤íŒ¨ (FAIL)"
            all_passed = False

        print(f"\n[í…ŒìŠ¤íŠ¸] ë²”ìœ„: {l} ~ {r}")
        print(f"  - ê¸°ëŒ€ê°’ (Expected): {expected_result}")
        print(f"  - ì‹¤ì œê°’ (Actual):   {actual_result}")
        print(f"  - ê²°ê³¼: {status}")

    print("\n-----------------------------------------------------------")
    if all_passed:
        print("ğŸ‰ ëª¨ë“  í…ŒìŠ¤íŠ¸ ì¼€ì´ìŠ¤ë¥¼ í†µê³¼í–ˆìŠµë‹ˆë‹¤.")
    else:
        print("âš ï¸ ì¼ë¶€ í…ŒìŠ¤íŠ¸ ì¼€ì´ìŠ¤ì—ì„œ ì˜¤ë¥˜ê°€ ë°œìƒí–ˆìŠµë‹ˆë‹¤. ì½”ë“œ ê²€í† ê°€ í•„ìš”í•©ë‹ˆë‹¤.")

# í…ŒìŠ¤íŠ¸ ì‹¤í–‰
if __name__ == "__main__":
    run_tests(test_cases)
```

* ê° í•™ìƒë“¤ì˜ ì„ ë°œ ê³ ì‚¬ ë“±ìˆ˜ë¥¼ ë‹´ì€ ì •ìˆ˜ ë°°ì—´ `rank`ì™€
* ì „êµ­ ëŒ€íšŒ ì°¸ì—¬ ê°€ëŠ¥ ì—¬ë¶€ê°€ ë‹´ê¸´ boolean ë°°ì—´ `attendance`

ì „êµ­ ëŒ€íšŒì— ì„ ë°œëœ í•™ìƒ ë²ˆí˜¸ë“¤ì„ **ë“±ìˆ˜ê°€ ë†’ì€ ìˆœì„œëŒ€ë¡œ** ê°ê° $a, b, c$ë²ˆì´ë¼ê³  í•  ë•Œ, ë‹¤ìŒ ìˆ˜ì‹ì„ return í•˜ëŠ” `solution` í•¨ìˆ˜ë¥¼ ì‘ì„±í•´ ì£¼ì„¸ìš”.

#### 1. ì…ì¶œë ¥ ì˜ˆ #1: `rank = [3, 7, 2, 5, 4, 6, 1]`, `attendance = [F, T, T, T, T, F, F]`

ëª¨ë“  í•™ìƒì´ ì°¸ì„ ê°€ëŠ¥í•©ë‹ˆë‹¤.

**ìµœì¢… ê³„ì‚°:** $10000 \times 0 + 100 \times 1 + 2 = 102$

#### 3. ì…ì¶œë ¥ ì˜ˆ #3: `rank = [6, 1, 5, 2, 3, 4]`, `attendance = [T, F, T, F, F, T]`

| 4 | **4** | **5** | True | **1ìˆœìœ„ (a=5)** |

**ìµœì¢… ê³„ì‚°:** $10000 \times 5 + 100 \times 2 + 0 = 50200$

```python
def solution2(rank, attendance):
    selected_students = sorted([
        (r, idx)
        for idx, (r, is_present) in enumerate(zip(rank, attendance))
        if is_present
    ])

    # ìƒìœ„ 3ëª…ì˜ í•™ìƒ ë²ˆí˜¸(idx) ì¶”ì¶œ
    a, b, c = selected_students[0][1], selected_students[1][1], selected_students[2][1]
    return 10000 * a + 100 * b + c

# ----------------------------------------------------------------------
## ğŸ“ ì…ì¶œë ¥ ì˜ˆì‹œ ì±„ì  êµ¬ì¡°

# ğŸ“ ì…ì¶œë ¥ ì˜ˆì‹œ ë°ì´í„°
# (ì…ë ¥ê°’: rank, attendance, ê¸°ëŒ€ê°’: result)
test_cases = [
    # í”„ë¡œê·¸ë˜ë¨¸ìŠ¤ ê¸°ë³¸ ì…ì¶œë ¥ ì˜ˆì‹œ #1
    ([3, 7, 2, 5, 4, 6, 1], [False, True, True, True, True, False, False], 20403),
    # 1ìˆœìœ„(2ë“±): í•™ìƒ2 (a=2)
    # 2ìˆœìœ„(4ë“±): í•™ìƒ4 (b=4)
    # 3ìˆœìœ„(5ë“±): í•™ìƒ3 (c=3)

    # í”„ë¡œê·¸ë˜ë¨¸ìŠ¤ ê¸°ë³¸ ì…ì¶œë ¥ ì˜ˆì‹œ #2
    ([1, 2, 3], [True, True, True], 102),
    # 1ìˆœìœ„(1ë“±): í•™ìƒ0 (a=0)
    # 2ìˆœìœ„(2ë“±): í•™ìƒ1 (b=1)
    # 3ìˆœìœ„(3ë“±): í•™ìƒ2 (c=2)

    # í”„ë¡œê·¸ë˜ë¨¸ìŠ¤ ê¸°ë³¸ ì…ì¶œë ¥ ì˜ˆì‹œ #3
    ([6, 1, 5, 2, 3, 4], [True, False, True, False, False, True], 50200),
    # 1ìˆœìœ„(2ë“±): í•™ìƒ3 (íƒˆë½)
    # 1ìˆœìœ„(4ë“±): í•™ìƒ5 (a=5)
    # 2ìˆœìœ„(5ë“±): í•™ìƒ2 (b=2)
    # 3ìˆœìœ„(6ë“±): í•™ìƒ0 (c=0)

    # ì¶”ê°€ í…ŒìŠ¤íŠ¸ ì¼€ì´ìŠ¤ 1: ë“±ìˆ˜ 1, 2, 3ë“±ì´ ëª¨ë‘ ì°¸ì„ ë¶ˆê°€ëŠ¥í•œ ê²½ìš°
    ([1, 2, 3, 4, 5, 6], [False, False, False, True, True, True], 30405),
    # 1ìˆœìœ„(4ë“±): í•™ìƒ3 (a=3)
    # 2ìˆœìœ„(5ë“±): í•™ìƒ4 (b=4)
    # 3ìˆœìœ„(6ë“±): í•™ìƒ5 (c=5)
]

def run_tests(test_cases):
    """ì£¼ì–´ì§„ ì…ì¶œë ¥ ì˜ˆì‹œë¡œ solution í•¨ìˆ˜ë¥¼ í…ŒìŠ¤íŠ¸í•˜ê³  ê²°ê³¼ë¥¼ ì¶œë ¥í•©ë‹ˆë‹¤."""
    print("--- ğŸ† í”„ë¡œê·¸ë˜ë¨¸ìŠ¤ 'ì „êµ­ ëŒ€íšŒ ì„ ë°œ ê³ ì‚¬' í…ŒìŠ¤íŠ¸ ì‹¤í–‰ ---")

    all_passed = True

    for rank, attendance, expected_result in test_cases:
        actual_result = solution2(rank, attendance)

        # ê²°ê³¼ ë¹„êµ ë° ì¶œë ¥
        if actual_result == expected_result:
            status = "âœ… í†µê³¼ (PASS)"
        else:
            status = "âŒ ì‹¤íŒ¨ (FAIL)"
            all_passed = False

        print(f"\n[í…ŒìŠ¤íŠ¸] rank ê¸¸ì´: {len(rank)}")
        print(f"  - ê¸°ëŒ€ê°’ (Expected): {expected_result}")
        print(f"  - ì‹¤ì œê°’ (Actual):   {actual_result}")
        print(f"  - ê²°ê³¼: {status}")

    print("\n-------------------------------------------------------------")
    if all_passed:
        print("ğŸ‰ ëª¨ë“  í…ŒìŠ¤íŠ¸ ì¼€ì´ìŠ¤ë¥¼ í†µê³¼í–ˆìŠµë‹ˆë‹¤.")
    else:
        print("âš ï¸ ì¼ë¶€ í…ŒìŠ¤íŠ¸ ì¼€ì´ìŠ¤ì—ì„œ ì˜¤ë¥˜ê°€ ë°œìƒí–ˆìŠµë‹ˆë‹¤. ì½”ë“œ ê²€í† ê°€ í•„ìš”í•©ë‹ˆë‹¤.")

# í…ŒìŠ¤íŠ¸ ì‹¤í–‰
if __name__ == "__main__":
    run_tests(test_cases)
```

## ğŸ“ ë¬¸ì œ ì„¤ëª…

### ì…ì¶œë ¥ ì˜ˆ ì„¤ëª… (ì˜ˆì œ #1)

ì˜ˆì œ 1ë²ˆì˜ `my_string`ì—ì„œ ì¸ë±ìŠ¤ 2ë¶€í„° `overwrite_string`ì˜ ê¸¸ì´(7)ë§Œí¼ì— í•´ë‹¹í•˜ëŠ” ë¶€ë¶„ì€ "11oWor1"ì…ë‹ˆë‹¤. ì´ë¥¼ "lloWorl"ë¡œ ë°”ê¾¼ "HelloWorld"ë¥¼ return í•©ë‹ˆë‹¤.

### ì…ì¶œë ¥ ì˜ˆ ì„¤ëª… (ì˜ˆì œ #2)

ì˜ˆì œ 2ë²ˆì˜ `my_string`ì—ì„œ ì¸ë±ìŠ¤ 7ë¶€í„° `overwrite_string`ì˜ ê¸¸ì´(7)ë§Œí¼ì— í•´ë‹¹í•˜ëŠ” ë¶€ë¶„ì€ "29b8UYP"ì´ê³  ì´ë¥¼ "merS123"ë¡œ ë°”ê¾¼ "ProgrammerS123"ë¥¼ return í•©ë‹ˆë‹¤.

```python
def solution3(my_string, overwrite_string, s):
    char_list = list(my_string)
    char_list[s : s + len(overwrite_string)] = list(overwrite_string)
    return "".join(char_list)

# ----------------------------------------------------------------------
## ğŸ“ ì…ì¶œë ¥ ì˜ˆì‹œ ì±„ì  êµ¬ì¡°

# ğŸ“ ì…ì¶œë ¥ ì˜ˆì‹œ ë°ì´í„°
# (ì…ë ¥ê°’: my_string, overwrite_string, s, ê¸°ëŒ€ê°’: result)
test_cases = [
    # í”„ë¡œê·¸ë˜ë¨¸ìŠ¤ ê¸°ë³¸ ì…ì¶œë ¥ ì˜ˆì‹œ #1
    ("He11oWor1d", "lloWorl", 2, "HelloWorld"),

    # í”„ë¡œê·¸ë˜ë¨¸ìŠ¤ ê¸°ë³¸ ì…ì¶œë ¥ ì˜ˆì‹œ #2
    ("Program29b8UYP", "merS123", 7, "ProgrammerS123"),

    # ì¶”ê°€ í…ŒìŠ¤íŠ¸ ì¼€ì´ìŠ¤ 1: s=0 (ì‹œì‘ë¶€í„° ê²¹ì³ì“°ê¸°)
    ("abcde", "XYZ", 0, "XYZde"),

    # ì¶”ê°€ í…ŒìŠ¤íŠ¸ ì¼€ì´ìŠ¤ 2: ëê¹Œì§€ ê²¹ì³ì“°ê¸° (ì œí•œì‚¬í•­: ê²¹ì³ì“°ê¸° ê¸¸ì´ <= ì›ë³¸ ê¸¸ì´)
    ("ABCDEFGHI", "xyz", 6, "ABCDEFxyz"), # ì›ë³¸ ê¸¸ì´ 9, ê²¹ì³ì“°ê¸° ê¸¸ì´ 3, s=6. 6+3=9(ë)

    # ì¶”ê°€ í…ŒìŠ¤íŠ¸ ì¼€ì´ìŠ¤ 3: ì „ì²´ ê²¹ì³ì“°ê¸°
    ("Python", "Rust", 0, "Ruston"), # ì œí•œì‚¬í•­ ìƒ ì „ì²´ë¥¼ ê²¹ì³ì“°ëŠ” ê²½ìš°ëŠ” s=0ì´ê³  ê¸¸ì´ê°€ ê°™ì„ ë•Œë§Œ ê°€ëŠ¥
]

def run_tests(test_cases):
    """ì£¼ì–´ì§„ ì…ì¶œë ¥ ì˜ˆì‹œë¡œ solution í•¨ìˆ˜ë¥¼ í…ŒìŠ¤íŠ¸í•˜ê³  ê²°ê³¼ë¥¼ ì¶œë ¥í•©ë‹ˆë‹¤."""
    print("--- ğŸ“ í”„ë¡œê·¸ë˜ë¨¸ìŠ¤ 'ë¬¸ìì—´ ê²¹ì³ì“°ê¸°' í…ŒìŠ¤íŠ¸ ì‹¤í–‰ ---")

    all_passed = True

    for my_string, overwrite_string, s, expected_result in test_cases:
        actual_result = solution3(my_string, overwrite_string, s)

        # ê²°ê³¼ ë¹„êµ ë° ì¶œë ¥
        if actual_result == expected_result:
            status = "âœ… í†µê³¼ (PASS)"
        else:
            status = "âŒ ì‹¤íŒ¨ (FAIL)"
            all_passed = False

        print(f"\n[í…ŒìŠ¤íŠ¸] s={s}, ê²¹ì³ì“¸ ë¬¸ìì—´: \"{overwrite_string}\"")
        print(f"  - ì›ë³¸:      \"{my_string}\"")
        print(f"  - ê¸°ëŒ€ ê²°ê³¼: \"{expected_result}\"")
        print(f"  - ì‹¤ì œ ê²°ê³¼: \"{actual_result}\"")
        print(f"  - ê²°ê³¼: {status}")

    print("\n-----------------------------------------------------------")
    if all_passed:
        print("ğŸ‰ ëª¨ë“  í…ŒìŠ¤íŠ¸ ì¼€ì´ìŠ¤ë¥¼ í†µê³¼í–ˆìŠµë‹ˆë‹¤.")
    else:
        print("âš ï¸ ì¼ë¶€ í…ŒìŠ¤íŠ¸ ì¼€ì´ìŠ¤ì—ì„œ ì˜¤ë¥˜ê°€ ë°œìƒí–ˆìŠµë‹ˆë‹¤. ì½”ë“œ ê²€í† ê°€ í•„ìš”í•©ë‹ˆë‹¤.")

# í…ŒìŠ¤íŠ¸ ì‹¤í–‰
if __name__ == "__main__":
    run_tests(test_cases)
```

## ğŸ’¡ ë¬¸ì œ ì„¤ëª…

ì •ìˆ˜ ë°°ì—´ `arr`ì™€ 2ì°¨ì› ì •ìˆ˜ ë°°ì—´ `queries`ê°€ ì£¼ì–´ì§‘ë‹ˆë‹¤.

`queries`ì˜ ì›ì†ŒëŠ” ê°ê° í•˜ë‚˜ì˜ queryë¥¼ ë‚˜íƒ€ë‚´ë©°, `[s, e, k]` ê¼´ì…ë‹ˆë‹¤.

ê° queryë§ˆë‹¤ ìˆœì„œëŒ€ë¡œ $s \le i \le e$ ì¸ ëª¨ë“  $i$ì— ëŒ€í•´ **$k$ë³´ë‹¤ í¬ë©´ì„œ ê°€ì¥ ì‘ì€ $\text{arr}[i]$**ë¥¼ ì°¾ìŠµë‹ˆë‹¤.

ê° ì¿¼ë¦¬ì˜ ìˆœì„œì— ë§ê²Œ ë‹µì„ ì €ì¥í•œ ë°°ì—´ì„ ë°˜í™˜í•˜ëŠ” `solution` í•¨ìˆ˜ë¥¼ ì™„ì„±í•´ ì£¼ì„¸ìš”.

ë‹¨, íŠ¹ì • ì¿¼ë¦¬ì˜ ë‹µì´ ì¡´ì¬í•˜ì§€ ì•Šìœ¼ë©´ **-1**ì„ ì €ì¥í•©ë‹ˆë‹¤.

---

## ğŸ”’ ì œí•œì‚¬í•­

* $1 \le \text{arr}$ì˜ ê¸¸ì´ $\le 1,000$
* $0 \le \text{arr}$ì˜ ì›ì†Œ $\le 1,000,000$
* $1 \le \text{queries}$ì˜ ê¸¸ì´ $\le 1,000$
* $0 \le s \le e < \text{arr}$ì˜ ê¸¸ì´
* $0 \le k \le 1,000,000$

---

## ğŸ“ ì…ì¶œë ¥ ì˜ˆ

| arr | queries | result |
| :---: | :---: | :---: |
| `[0, 1, 2, 4, 3]` | `[[0, 4, 2],[0, 3, 2],[0, 2, 2]]` | `[3, 4, -1]` |

### ì…ì¶œë ¥ ì˜ˆ ì„¤ëª… (ì˜ˆì œ #1)

* **ì›ë³¸ ë°°ì—´ `arr`:** `[0, 1, 2, 4, 3]`

| ì¿¼ë¦¬ | êµ¬ê°„ [s, e] | ëª©í‘œê°’ (k) | êµ¬ê°„ ë‚´ ì›ì†Œ (arr[i]) | k(2)ë³´ë‹¤ í° ê°’ | ê°€ì¥ ì‘ì€ ê°’ | ê²°ê³¼ |
| :---: | :---: | :---: | :---: | :---: | :---: | :---: |
| 1 | [0, 4] | 2 | 0, 1, 2, 4, 3 | 4, 3 | **3** | 3 |
| 2 | [0, 3] | 2 | 0, 1, 2, 4 | 4 | **4** | 4 |
| 3 | [0, 2] | 2 | 0, 1, 2 | ì—†ìŒ | - | **-1** |

ë”°ë¼ì„œ `[3, 4, -1]`ì„ return í•©ë‹ˆë‹¤.

```python
def solution4(arr, queries):
    result = []
    for s, e, k in queries:
        candidates = [x for x in arr[s : e + 1] if x > k]
        result.append(min(candidates) if candidates else -1)

    return result

# ----------------------------------------------------------------------
## ğŸ“ ì…ì¶œë ¥ ì˜ˆì‹œ ì±„ì  êµ¬ì¡°

# ğŸ“ ì…ì¶œë ¥ ì˜ˆì‹œ ë°ì´í„°
# (ì…ë ¥ê°’: arr, queries, ê¸°ëŒ€ê°’: result)
test_cases = [
    # í”„ë¡œê·¸ë˜ë¨¸ìŠ¤ ê¸°ë³¸ ì…ì¶œë ¥ ì˜ˆì‹œ
    ([0, 1, 2, 4, 3], [[0, 4, 2], [0, 3, 2], [0, 2, 2]], [3, 4, -1]),

    # ì¶”ê°€ í…ŒìŠ¤íŠ¸ ì¼€ì´ìŠ¤ 1: êµ¬ê°„ì´ ì§§ê³  k ê°’ì´ ë‹¤ì–‘í•œ ê²½ìš°
    ([10, 5, 20, 15, 30], [[0, 2, 15], [1, 4, 5], [3, 4, 25]], [20, 15, 30]),
    # Q1 [10, 5, 20], k=15: 20 -> 20
    # Q2 [5, 20, 15, 30], k=5: 15 -> 15 (ì£¼ì˜: 20ë„ ìˆì§€ë§Œ, 15ê°€ ë” ì‘ìŒ) -> ì •ì •: 15
    # Q3 [15, 30], k=25: 30 -> 30

    # ì¶”ê°€ í…ŒìŠ¤íŠ¸ ì¼€ì´ìŠ¤ 2: kë³´ë‹¤ í° ê°’ì´ ì—†ëŠ” ê²½ìš°
    ([1, 1, 1, 1, 1], [[0, 4, 1]], [-1]),

    # ì¶”ê°€ í…ŒìŠ¤íŠ¸ ì¼€ì´ìŠ¤ 3: ëª¨ë“  ê°’ì´ kë³´ë‹¤ í° ê²½ìš°
    ([100, 200, 300], [[0, 2, 50]], [100]),
]

def run_tests(test_cases):
    """ì£¼ì–´ì§„ ì…ì¶œë ¥ ì˜ˆì‹œë¡œ solution í•¨ìˆ˜ë¥¼ í…ŒìŠ¤íŠ¸í•˜ê³  ê²°ê³¼ë¥¼ ì¶œë ¥í•©ë‹ˆë‹¤."""
    print("--- ğŸ”¢ í”„ë¡œê·¸ë˜ë¨¸ìŠ¤ 'ìˆ˜ì—´ê³¼ êµ¬ê°„ ì¿¼ë¦¬ 2' í…ŒìŠ¤íŠ¸ ì‹¤í–‰ ---")

    all_passed = True

    for arr, queries, expected_result in test_cases:
        actual_result = solution4(arr, queries)

        # ê²°ê³¼ ë¹„êµ ë° ì¶œë ¥
        if actual_result == expected_result:
            status = "âœ… í†µê³¼ (PASS)"
        else:
            status = "âŒ ì‹¤íŒ¨ (FAIL)"
            all_passed = False

        print(f"\n[í…ŒìŠ¤íŠ¸] arr: {arr}")
        print(f"  - ê¸°ëŒ€ê°’ (Expected): {expected_result}")
        print(f"  - ì‹¤ì œê°’ (Actual):   {actual_result}")
        print(f"  - ê²°ê³¼: {status}")

    print("\n-----------------------------------------------------------")
    if all_passed:
        print("ğŸ‰ ëª¨ë“  í…ŒìŠ¤íŠ¸ ì¼€ì´ìŠ¤ë¥¼ í†µê³¼í–ˆìŠµë‹ˆë‹¤.")
    else:
        print("âš ï¸ ì¼ë¶€ í…ŒìŠ¤íŠ¸ ì¼€ì´ìŠ¤ì—ì„œ ì˜¤ë¥˜ê°€ ë°œìƒí–ˆìŠµë‹ˆë‹¤. ì½”ë“œ ê²€í† ê°€ í•„ìš”í•©ë‹ˆë‹¤.")

# í…ŒìŠ¤íŠ¸ ì‹¤í–‰
if __name__ == "__main__":
    run_tests(test_cases)
```
